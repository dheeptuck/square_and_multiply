```python
def SquareMultiply(value, power):
    result=1

    while power>0:
        if power & 0x01:
            result = result * value
        value = value*value
        power = power>>1
    return result
```

```python
def BruteForcePower(value, power):
    result=1
    for i in range(power):
        result = result*value
    return result
```

```python
import time

def ComputeTime(fx, arg1, arg2):
    init_time = time.time()
    fx(arg1, arg2)
    return time.time() - init_time
```

```python
import pylab as pl

x = [x for x in range(1,30000,10)]
y_bruteforce = [ComputeTime(BruteForcePower, 3, i)*(10**6) for i in x]
y_optimized = [ComputeTime(SquareMultiply, 3, i)*(10**6) for i in x]

pl.plot(x, y_bruteforce, 'r.')
pl.plot(x, y_optimized, 'b.')
pl.xlabel('exponent', fontsize=10)
pl.ylabel('time(us)', fontsize=10)
pl.legend(["BruteForcePower", "SquareMultiply"], loc ="upper right")


pl.show()
```
